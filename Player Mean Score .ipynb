```python
import pandas as pd
import numpy as np

```

```python
train= pd.read_csv('team_avg_balls_played.csv',sep=',')

print(train.head(10))
```

```python
Players = train.groupby('Batsman')
print(Players.groups)
```

```python
Mean=Players['balls'].agg(np.mean)
Mean
```

```python
MeanTable = pd.DataFrame({'Mean':Mean}, index = Mean.index)

#Visualize the first 5 rows
MeanTable.head()
```

```python
filename = 'PlayerMeanScore.csv'

MeanTable.to_csv(filename,index=True)

print('Saved file: ' + filename)
```
