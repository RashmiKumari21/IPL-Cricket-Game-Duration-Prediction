```python
path = "/Users/swayam/Documents/Projects/Data_Mining/Data"
if os.path.isdir(path+"/Team/Rating") is False:
    os.mkdir(path+"/Team/Rating")
for year in years:
    if os.path.isdir("/Users/swayam/Documents/Projects/Data_Mining/Data/Team/Rating/" + year) is False:
        os.mkdir("/Users/swayam/Documents/Projects/Data_Mining/Data/Team/Rating/" + year)
    data = pd.read_csv("%s/IPL Website Scrapping/%s.csv"%(path,year))
    players = []
    for player in data['PLAYER']:
        if type(player) == str:
            players.append(player.replace(u'\xa0', ' '))
        else:
            players.append(player)
    data =  data.drop(['PLAYER'], axis = 1)
    data['PLAYER']  = players
    data.drop_duplicates(subset = 'PLAYER', keep ='last', inplace = True)
    data = data.set_index('PLAYER')
    data = data.drop_duplicates(keep ='first') 
    team_data = pd.read_csv("%s/Team/Data/%s.csv"%(path,year))
    for team in team_data.columns:
        team_year_by_year = pd.read_csv("%s/Team/Data_By_Year/%s/%s.csv"%(path,year,team))
        team_year_by_year = team_year_by_year.set_index(team)
        df = pd.DataFrame()
        for player in team_data[team]:
            if player in data.index:
                player_stat = data[data.index == player][data.columns[1:9]]
                df = pd.concat([df, player_stat], axis = 0)
            elif player in names and names[player] in data.index:
                as_list = data.index.tolist()
                idx = as_list.index(names[player])
                as_list[idx] = player
                data.index = as_list
                player_stat = data[data.index == player][data.columns[1:9]]
                df = pd.concat([df, player_stat], axis = 0)
        team_year_by_year = pd.concat([team_year_by_year, df], axis = 1, sort = True)
        team_year_by_year[team] = team_year_by_year.index
        col = ['Season', team, 'Pts', 'Mat', 'Wkts', 'Dots', '4s', '6s', 'Catches', 'Stumpings']
        team_year_by_year = team_year_by_year[col]
        team_year_by_year.to_csv("%s/Team/Rating/%s/%s.csv"%(path,year,team), index = False)
```
