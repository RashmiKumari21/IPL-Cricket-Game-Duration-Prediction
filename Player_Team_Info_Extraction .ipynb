```python
import pandas as pd
import os
from itertools import repeat
```

```python
player_team_info = pd.read_csv("/Users/swayam/Documents/Projects/Data_Mining/Data/Players_Team_info.csv")
teams = pd.read_csv("/Users/swayam/Documents/Projects/Data_Mining/Data/Teams.csv")
```

```python
dictionary = dict()
for column in player_team_info.columns[2:]:
    dictionary[column] = dict()
```

```python
for key in dictionary.keys():
    for team in teams.iterrows(): 
        if team[1]['Indicated by word'] in set(player_team_info[str(key)]):
            dictionary[key][team[1]['Team']] = player_team_info['Name'][player_team_info[key] == team[1]['Indicated by word']].values
```

```python
if os.path.isdir("/Users/swayam/Documents/Projects/Data_Mining/Data/Team/Data") is False:
    os.mkdir("/Users/swayam/Documents/Projects/Data_Mining/Data/Team/Data")
for year in dictionary.keys():
    df = pd.DataFrame()
    for team in dictionary[year].keys():
        #if os.path.isdir("/Users/swayam/Documents/Projects/Data_Mining/Data/Team/" + year) is False:
            #os.mkdir("/Users/swayam/Documents/Projects/Data_Mining/Data/Team/" + year)
        data =  pd.DataFrame({team: dictionary[year][team]})
        data.index.name='Season'
        df = pd.concat([df,data], axis = 1)
    #df.index.name='Season'
    df = df.reset_index(drop = True)
    df.to_csv("/Users/swayam/Documents/Projects/Data_Mining/Data/Team/Data/%s.csv"%(year), index = False)
        
```

```python
def team_team_info(team1, team2, year):
    import pandas as pd
    path = "/Users/swayam/Documents/Projects/Data_Mining/Data/Team/Data/%s.csv"%(year)
    data = pd.read_csv(path)
    team1 = data[team1]
    team2 = data[team2]
    
    teams = pd.concat([team1, team2], axis = 1)
    teams = teams.fillna('Player Name')
    
    return teams
```
