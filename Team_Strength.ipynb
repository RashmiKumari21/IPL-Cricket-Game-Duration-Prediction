```python
import pandas as pd
import os
```

```python
data_match = pd.read_csv('/Users/swayam/Documents/Projects/Data_Mining/Data/Match_Info.csv')
teams = pd.read_csv("/Users/swayam/Documents/Projects/Data_Mining/Data/Teams.csv").Team.values
path = ("/Users/swayam/Documents/Projects/Data_Mining/Data/Team/Rating")
```

```python
data_match = data_match.replace('Rising Pune Supergiants', 'Rising Pune Supergiant')
data_match = data_match.replace('Pune Warriors', 'Pune Warriors India')
index1 = data_match[data_match.Team1 == 'Delhi Capitals'].index
index2 = data_match[data_match.Team2 == 'Delhi Capitals'].index
delhi_capitals_index = list(index1) + list(index2)
data_match = data_match.replace('Delhi Capitals','Delhi Daredevils')
```

```python
data_match.iloc[delhi_capitals_index]
```

```python
years = ['2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
```

```python
result = dict()
team_apperance = dict()


for year in years:
    result[year] = dict()
    team_apperance[year] = dict()
        
for year in years:
    for team in teams:
        if os.path.isfile("%s/%s/%s.csv"%(path,year,team)) is True:
            data = pd.read_csv("%s/%s/%s.csv"%(path,year,team)).dropna()
            data = data.sort_values(by = ['Mat'], axis = 0, ascending = False)
            top_eleven = data.head(11)
            result[year][team] = top_eleven['Pts'].sum()
        else: 
            continue

total_points = pd.DataFrame(result)
total_points
```

```python
strength = pd.DataFrame()
for index, row in data_match.iterrows():
    #data = data_match[data_match.Season == row.Season]
    apperance = data_match[data_match['Date'] <= row.Date]
    team1 = row['Team1']
    team1_apperance = apperance[apperance.Team1 == team1].shape[0]
    team1_points = total_points[str(row.Season)][team1]
    team1_strength = team1_points / team1_apperance
    
    team2 = row['Team2']
    team2_apperance = apperance[apperance.Team2 == team2].shape[0]
    team2_points = total_points[str(row.Season)][team2]
    team2_strength = team2_points / team2_apperance
    
    #if index == 1:
     #   break
    temp = pd.DataFrame({'Team1_Strength':team1_strength, 'Team2_Strength': team2_strength}, index = [index])
    strength = pd.concat([strength, temp])
```

```python
data_match = pd.concat([data_match, strength], axis = 1)
data_match.to_csv('/Users/swayam/Documents/Projects/Data_Mining/Data/Team_Strength_Match_Info.csv', index = False)
data_match
```
